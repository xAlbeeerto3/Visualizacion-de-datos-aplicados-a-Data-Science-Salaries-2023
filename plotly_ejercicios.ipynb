#### Preguntas a resolver usando las librerías de creaciones gráficas vistas en clase: 

¿Cuál es el salario promedio por nivel de experiencia?

```python
salario_medio_df = pd.DataFrame(df)
fig = px.strip(salario_medio_df, x="experience_level", y="salary_in_usd", template="plotly_dark", color='job_title', title="Salario según nivel de experiencia", )
fig.update_xaxes(tickangle=90)
fig.show()
```

¿Cómo ha cambiado el salario medio con el paso del tiempo (work_year)?

```python
# Antes de hacer gráfica, tenemos que calcular el salario medio por año (Salary per year).

Salario_medio_por_año = df.groupby('work_year')['salary_in_usd'].mean().reset_index()
Salario_medio_por_año # Una vez conocemos este dato si podemos analizar su evolución.
```

```python
# Ejemplo utilizando Plotly
cambio_salario = pd.DataFrame(df)
fig = px.strip(cambio_salario, x="work_year", y="salary_in_usd", template="plotly_dark", color='job_title', title="Salario medio con el paso del tiempo", )
fig.update_xaxes(tickangle=90)
fig.show()
```

```python
# Otro ejemplo con Seaborn
salario_medio = sns.stripplot(x="work_year", y="salary_in_usd", data=df,jitter=True,palette='Set2',dodge=True)
salario_medio.set_xlabel('Años de experiencia laboral')
salario_medio.set_ylabel('Salario en USD ($)')
plt.show()  
```

¿Cuál es la proporción de trabajo remoto entre diferentes roles de trabajo?

```python
# Para este ejercicio, volvemos a calcular la media. En este caso se necesita la proporción media de trabajo remoto para cada puesto de trabajo.

prop_media_trabajo_remoto = df.groupby('job_title')['remote_ratio'].mean().reset_index()

fig = px.bar(
    prop_media_trabajo_remoto,
    x='job_title',  # Puestos de trabajo
    y='remote_ratio',  # Proporción media de trabajo remoto
    title='Proporción media de trabajo remoto para cada puesto',
    template="plotly_dark",
    color_discrete_sequence=['yellow'],
    height=1200,
    orientation='v')

fig.show()
```

¿Existe una relación entre el tamaño de la empresa y el salario?

```python
fig = px.box(df, x='company_size', y='salary_in_usd', 
             title='Relación tamaño de la empresa - salario', 
             category_orders={"company_size": ["S", "M", "L"]}, template="presentation", color='company_size')

fig.show()
```

```python
# También es interesante dar valores a S,M y L para calcular correlaciones.

df['company_size_num'] = df['company_size'].map({'S': 1, 'M': 2, 'L': 3})

# Calculamos la correlacion y creamos un mapa de calor
correlacion = df[['salary_in_usd', 'company_size_num']].corr()

# Usamos como ejemplo Pairplot
sns.pairplot(correlacion, diag_kind='kde')
plt.show()
```

¿Cuál es el país con más trabajadores remotos?

```python
# Necesitamos conocer la cantidad de trabajo remoto por pais

trabajos_remoto_por_pais = df.groupby('employee_residence')['remote_ratio'].sum().reset_index()

trabajos_remoto_por_pais
```

```python
# Ahora si creamos el grafico usando Catplot.
sns.set_theme(style='whitegrid')
plt.figure(figsize=(5, 5))
sns.catplot(data=trabajos_remoto_por_pais, x='employee_residence', y='remote_ratio', kind='point', height=6, aspect=3, palette='deep')
plt.title('Cantidad de trabajadores remotos por país')
plt.xticks(rotation=90)
plt.xlabel('Paises')
plt.ylabel('Total de trabajadores remotos')
plt.show()
```

¿Cómo ha cambiado la proporción de trabajo remoto con el paso del tiempo?

```python
# Igual pero con años en lugar de paises.
trabajos_remotos_por_años = df.groupby('work_year')['remote_ratio'].mean().reset_index()

fig = px.bar(trabajos_remotos_por_años, x='work_year', y='remote_ratio', color='remote_ratio')
fig.update_layout(title='Proporción de empleados remotos con el paso del tiempo')
fig.show()
```

¿Cuál es el país con los salarios promedio más altos en USD?

```python
# Calculamos el salario promedio por país.
salario_medio_por_pais = df.groupby('employee_residence')['salary_in_usd'].mean().reset_index()

fig = px.line(salario_medio_por_pais.sort_values('salary_in_usd', ascending=False),
              x='employee_residence', y='salary_in_usd', 
              title='Países con mayores salarios en $',
              template="seaborn", 
              color_discrete_sequence=['#FF6692'])

fig.show()
```

¿Cómo se distribuyen los roles de trabajo en las diferentes categorías de experiencia?

```python

rol_trabajo_experiencia = df.groupby(['experience_level', 'job_title']).size().reset_index(name='Count') 

fig = px.bar(rol_trabajo_experiencia, x="experience_level", y="job_title", color="job_title", title="Long-Form Input")
fig.show()
```

¿Cómo varía el salario promedio entre los diferentes tipos de empleo, fulltime, partime ect?

```python
salario_medio_segun_tipos_empleo = df.groupby('employment_type')['salary_in_usd'].mean().reset_index()

fig = px.bar(salario_medio_segun_tipos_empleo, x='employment_type', y='salary_in_usd',
             hover_data=['employment_type', 'salary_in_usd'], color='salary_in_usd',
             labels={'sal':'Salario en USD'}, height=400)
fig.show()
```

#### Crear un top10 con los empleos mas demandados:

```python
top_10_empleos_demandados = df.job_title.value_counts()[:10].index.tolist()

sns.countplot(data=df[df.job_title.isin(top_10_empleos_demandados)], y='job_title', palette='Set2')

plt.title('Top 10 empleos más demandados')
plt.xlabel('Salario en (USD)')
plt.ylabel('Título del empleo')
```

####  Crear un top10 con los empleos mas cotizados:

```python
top_10_empleos_cotizados = df.groupby('job_title')['salary_in_usd'].mean().reset_index().head(10).sort_values(by='salary_in_usd',ascending=False)

plt.figure(figsize=(10, 6))
sns.set_theme(style='whitegrid')

sns.barplot(data=top_10_empleos_cotizados, y='job_title', x='salary_in_usd', palette='Set3')

plt.title('Top 10 empleos más cotizados')
plt.xlabel('Salario medio (USD)')
plt.ylabel('Título del empleo')


plt.show
```

####  Crear un top10 con empleos donde podemos trabajar remoto: 

```python
top_10_empleos_remotos = df.groupby('job_title')['remote_ratio'].mean().reset_index().head(10).sort_values(by='remote_ratio',ascending=False)


plt.figure(figsize=(10, 6))
sns.set_theme(style='whitegrid')

sns.barplot(data=top_10_empleos_remotos, y='job_title', x='remote_ratio', palette='Set3')

plt.title('Top 10 empleos remotos')
plt.xlabel('Remote ratio')
plt.ylabel('Título del empleo')


plt.show
```
